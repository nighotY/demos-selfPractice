```python
import requests
import json
from pprint import pp
import re
```

```python
url = "https://swapi.dev/api/starships/?page=1"
data=[]
while url:
    response=requests.get(url)
    page_data=response.json()
    data.extend(page_data['results'])
    url=page_data['next']

print(len(data))
pp(data)
```

1. Write a function that returns all of the models of starships

```python
def get_models():
    models=[]
    for record in data:
        models.append(record['model'])
    return models

print(get_models())
```

2. Write a function that finds the ship that can carry the most cargo

```python
def max_cargo():
    cargo_capacity = {}
    for record in data:
        if record['cargo_capacity'].isnumeric():
            cargo_capacity[record['name']]=int(record['cargo_capacity'])
    sorted_cargo= sorted(cargo_capacity.items(),key=lambda x:x[1],reverse=True)
    return(sorted_cargo[0][0])

print(max_cargo())
```

3. Write a function that finds the number of crew and passengers on a given 
   starship

```python
def numberof_pass_crew(name):
    starship = {}
    for record in data:
        if record['name'] == name:
            starship['crew']=record['crew']
            starship['passanger']=record['passengers']
    return starship
```

```python
starship_name = "Star Destroyer"
result=numberof_pass_crew(starship_name)
print(f'{starship_name} has {result}')
```

4. Write a function to find the most expensive starship

```python
def most_expensive():
    starship_price = {}
    for record in data:
        if record['cost_in_credits'].isnumeric():
           starship_price[record['name']] = int(record['cost_in_credits'])
    starship_sort= sorted(starship_price.items(),key=lambda x:x[1],reverse=True)
    return starship_sort[0][0]

print(most_expensive())
```

5. Write a function to find all of the starships that are less than a given 
   price

```python
def starships_below(price):
    starships=[]
    for record in data:
        if record['cost_in_credits'].isnumeric():
            if int(record['cost_in_credits']) < price:
                starships.append(record['name'])
    return starships

print(starships_below(240000))
```

6. Write a function to find the starships that have appeared in a given number 
   of films

```python
def starship_movie(x):
    starships_by_movie = []
    for record in data:
        if len(record['films']) == x:
            starships_by_movie.append(record['name'])
    return starships_by_movie
print(starship_movie(3))
```

7. Write a function that finds the shortest and longest ship. Return should be 
   the names of the ship in addition to the numeric values. Convert values into feet

```python
def find_short_long():
    ships= {}
    for record in data:
        length=re.sub(",","",record['length'])
        ships[record['name']] = round(float(re.sub(",","",record['length']))/30,2)
    sorted_ships=sorted(ships.items(), key = lambda x:x[1])
    return sorted_ships[0], sorted_ships[-1]
shortest,longest = find_short_long()
print(f'Shortest starship = {shortest[0]}, lenght= {shortest[1]}')
print(f'Longest starship = {longest[0]}, lenght= {longest[1]}')
```
