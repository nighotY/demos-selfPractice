Task Let's get data for all the pages  using while loop:
    Psuedo-Code
    !) we will need list to store data
    1) we will start from page 1 but we do not know how many pages are there,
        but if we do know that in our response page we have a key "next" which gives 
        link to next page. and if it is null we do not have next page
        we will use flag to decide if we are going to continue in while loop
        we have to make requst to api to get data
        store data
        check "next " key to see if we have next page
        if yes increment our page count by one
        else set our flag to False




```python
baseurl = baseurl = "https://swapi.dev/api/people/?page="
data=[]
flag = True
page = 1

while flag:
    response = requests.get(baseurl + str(page))
    response_data=response.json()
    if response_data['next'] != None:
         page=page+1
         print(page)
         data.append(response_data)
    else:
        flag=False
        data.append(response_data)

pp(data)
```

Task 2)Create a function that counts the total number of males and females. Return both numbers
        -Let's create function gender_count() which will return male_count and female_count
        - we will need 2 counter one to count number of males(male_count) and the other one for females(female_count)
        -we will need to iterate over complete list and check for "gender" key value and incremnt male_count and female_count variable    respectively
        return male_count and female_count as tuple

```python
def gender_count():
    male_count=0 
    female_count=0
    for rec in clean_data:
        if rec["gender"] == "female":
            female_count += 1
        elif rec["gender"] == "male":
            male_count += 1
    return male_count,female_count
```

```python
Male_count,Female_count = gender_count()
print(f'Total number of males: {Male_count} \nTotal number of females: {Female_count}')
```

Challenge 2: For problem number 4, convert the centimeters to feet

Psuedo-Code
-lets create (height)list and store int value of height in the list
  but there are some records in clean_data list which has no numeric values 
  so we will need to check that first before appending height to our list
- To get min max sort the list in asending order. so index [0] of list our min value and index [-1] is max value
-To calculate avg sum all the elements of the height list and divide that by len of the list
  for challenge 2 convert min, max avg into feet = cm/30.48
-return min max and avg value 


```python
def get_min_max_avg():
    height = []
    min = 0
    max = 0
    avg = 0
    total = 0
    for record in clean_data:
         if record["height"].isnumeric():
            height.append(int(record['height']))
            total+=int(record['height'])
    height.sort()
    min = round(height[0]/30.48,2)
    max = round(height[-1]/30.48,2)
    avg=round(total/len(height),2)
    return min, max, avg
    
```

Let's call the function get_min_max_avg() and unpack values into min, max and avg variable and then print out values

```python
min_height, max_height, avg_height= get_min_max_avg()
print(f'Height in Feet --- \nMin = {min_height}, Max = {max_height} and Avg = {avg_height}')
```

Create a function that accepts two arguements: eye color and hair color. Return a dictionary with the min, max, and avg height based on the 
parameters.

```python
def get_min_max_avg_height(eyecolor,haircolor):
    height = []
    min = 0
    max = 0
    avg = 0
    
    result = {}
    for record in clean_data:
        if record["eye_color"] == eyecolor and record["hair_color"] == haircolor:
            if record["height"].isnumeric():
                height.append(int(record["height"]))
    height.sort()
    if len(height) > 0:
        result["min"] = height[0]
        result["max"] = height[-1]
        tot=0
        for i in height:
            tot+=i
        result["avg"] = round(tot/len(height),2)
    else:
        result["min"] = 0
        result["max"] = 0
        result["avg"] = 0
    return result

```
